```python

# Function to process the column
def process_temperature(value):
    if pd.isna(value):  # Check for NaN
        return None
    try:
        if 'K' in value and '°C' not in value:  # If it has 'K' but not '°C'
            return int(value.replace('K', '').strip())
        elif '°C' in value:  # If it has '°C', convert to Kelvin
            temp_celsius = float(value.replace('°C', '').strip())
            return temp_celsius + 273.15  # Convert to Kelvin
        else:  # For other cases, remove non-numeric characters
            return int(''.join(filter(str.isdigit, value)))
    except ValueError:  # Handle unexpected cases
        return None


# Apply the function to the column
merged_data12['Tc12_curie_cleaned'] = merged_data12['Tc12_curie'].apply(process_temperature)




```
