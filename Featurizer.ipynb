```python
def extract_elemental_fractions(comp):
    try:
        composition = Composition(comp)
        total = sum(composition.values())
        return {element: amount / total for element, amount in composition.items()}
    except:
        return {}
```

```python
import re
import pandas as pd
import numpy as np

def extract_elements(formula):
    """
    Given a chemical formula string (e.g., 'Pu3Sc'),
    extract a list of element symbols (e.g., ['Pu', 'Sc']).
    This simple regex assumes element symbols start with an uppercase letter
    possibly followed by a lowercase letter.
    """
    pattern = r'[A-Z][a-z]?'
    return re.findall(pattern, formula)

def impute_homo_element(row, df, default_mode=None):
    """
    Impute the HOMO_element for a given row, based on known rows in the DataFrame.
    The imputation is constrained to only consider candidate elements present in the material.

    Parameters:
      row: The current row (a Series) for which to impute HOMO_element.
      df: The full DataFrame.
      default_mode: A fallback value to use if no candidate can be imputed.

    Returns:
      The imputed HOMO_element value.
    """
    # If HOMO_element is already provided, keep it.
    if pd.notnull(row['HOMO_element']):
        return row['HOMO_element']

    # Extract candidate elements from the material formula.
    candidate_elements = extract_elements(row['formula'])

    # Debug print (optional): print candidate_elements
    # print(f"Material: {row['material']} --> Candidate Elements: {candidate_elements}")

    # Filter the DataFrame to rows with non-missing HOMO_element that are in the candidate list.
    subset = df[df['HOMO_element'].isin(candidate_elements)]

    if not subset.empty:
        # Compute the mode (most frequent value) among the candidate rows.
        mode_series = subset['HOMO_element'].mode()
        if not mode_series.empty:
            return mode_series.iloc[0]

    # If no candidate row is found, return default_mode or leave as NaN.
    return default_mode if default_mode is not None else np.nan

# --- Example Usage ---



# Apply the imputation function to each row.
df['HOMO_element_imputed'] = df.apply(lambda row: impute_homo_element(row, df), axis=1)


```

```python

def get_mode(series):
    modes = series.mode()
    if not modes.empty:
        return modes.iloc[0]
    return None

# Group by HOMO_element_imputed and compute the mode for HOMO_character in each group.
element_to_mode = (
    df[df['HOMO_character'].notnull()]
    .groupby('HOMO_element_imputed')['HOMO_character']
    .agg(get_mode)
    .to_dict()
)




def impute_homo_character(row, mapping):
    # If HOMO_character is already known, keep it.
    if pd.notnull(row['HOMO_character']):
        return row['HOMO_character']

    # Get the element from HOMO_element_imputed.
    element = row['HOMO_element_imputed']

    # Look up the mode for that element in the mapping.
    # If no mode is available, leave as NaN or set a default.
    return mapping.get(element, np.nan)

# Apply the imputation function row-by-row.
df['HOMO_character_imputed'] = df.apply(lambda row: impute_homo_character(row, element_to_mode), axis=1)

```
